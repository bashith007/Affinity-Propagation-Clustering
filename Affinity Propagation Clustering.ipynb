```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_customers.csv")
dataset
```

```python
x = dataset.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import AffinityPropagation
list1=[]
for i in range(1, 11):
    clustering = AffinityPropagation(damping=0.9,convergence_iter=i,random_state=5,)
    clustering.fit(x)
    list1.append(clustering.labels_)
plt.plot(range(1, 11), list1)
plt.title("the line Method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
from sklearn.cluster import AffinityPropagation
clustering = AffinityPropagation(damping=0.9,convergence_iter=3,random_state=5,)
y_clustering = clustering.fit_predict(x)
y_clustering
```

```python
supervised=pd.DataFrame(dataset)
```

```python
supervised["Cluster_group"]=y_clustering
supervised
```

```python
supervised.to_csv("cluster.csv",index=False)
```
